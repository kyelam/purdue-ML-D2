# matplotlib

```python
import matplotlib.pyplot as plt

%matplotlib inline # magic commands
```

```python
import numpy as np

x = np.linspace(0,5,11)

x
```

```python
y = x**2
y
```

##### basic matplotlib commands

```python
plt.plot(x,y,'r');
plt.xlabel('X Axis title')
plt.ylabel('Y Axis title')
plt.title('Plot title')
plt.show()
```

##### create subplots

```python
plt.subplot(1,4,1) # nrows, ncols, plot_number

plt.plot(x,y,'r')

plt.subplot(1,4,2) # nrows, ncols, plot_number

plt.plot(y,x,'g')

plt.subplot(1,4,3) # nrows, ncols, plot_number

plt.plot(x,y*x,'b')

plt.subplot(1,4,4) # nrows, ncols, plot_number

plt.plot(x**2,y**3,'r')
```
