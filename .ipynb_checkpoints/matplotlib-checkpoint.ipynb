```python
import matplotlib.pyplot as plt

%matplotlib inline 
# magic commands
```

```python
import numpy as np

x = np.linspace(0,5,11)

x
```

```python
y = x**2
y
```

##### basic matplotlib commands

```python
plt.plot(x,y,'r');
plt.xlabel('X Axis title')
plt.ylabel('Y Axis title')
plt.title('Plot title')
plt.show()
```

##### create subplots

```python
plt.subplot(1,4,1) # nrows, ncols, plot_number

plt.plot(x,y,'r')

plt.subplot(1,4,2) # nrows, ncols, plot_number

plt.plot(y,x,'g')

plt.subplot(1,4,3) # nrows, ncols, plot_number

plt.plot(x,y*x,'b')

plt.subplot(1,4,4) # nrows, ncols, plot_number

plt.plot(x**2,y**3,'r')
```

#### Object oriented method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, height, width

axes.plot(x,y,'r')

axes.set_xlabel('X axis title')

axes.set_ylabel('Y axis title')

axes.set_title("Plot Title")
```

##### Legends

```python
fig = plt.figure()#figsize(15,8))
#fig.figsize(15,8)

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x**2,label="x**2")
axes.plot(x, x**3,label="x**3")

axes.legend()#loc=3)
```

```python
fig, ax = plt.subplots()

ax.plot(x,x**2,'b.-')
ax.plot(x,x**3,'g--')
ax.plot(x,x*y*10,color="#800080",alpha=0.5,linewidth=5,linestyle='-.')
ax.plot(y,x*y*10,color="#800080",alpha=0.5,linewidth=5,linestyle=':')
```

```python
fig, ax = plt.subplots()

ax.plot(y,x*y*5,
        color="#FF00FF",
        alpha=0.5,
        linewidth=5,
        linestyle='-',
        marker="P",
        markersize=16,
        markerfacecolor="blue",
        markeredgecolor="red",markeredgewidth=2);
```
